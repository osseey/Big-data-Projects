```python
# 1) Chargez le fichier dans un DataFrame
def load_data():
    df = pd.read_csv("couverture-2g-3g-4g-en-france-par-operateur-juillet-2015.csv",sep=";", header=None, dtype=object)
    headers = df.iloc[0]
    df1  = pd.DataFrame(df.values[1:], columns=headers)
    #df1.melt(id_vars=df1.columns[:6])
    return df1
```

```python
# load_data
load_data()
#print(df.shape)
```

```python
# 2) Transformez ce DataFrame large en un DataFrame long
#(1097820, 9)
def data_wide2long():
    df=load_data()
    liste=df.columns[:6]#.append("coordonnees")
    liste=['CODE POSTAL', 'CODE INSEE', 'NOM COMMUNE', 'CODE DEPARTEMENT',
       'SURFACE COMMUNE (km²)', 'POPULATION COMMUNE','coordonnees']
    df=df.melt(id_vars=liste,var_name="variable").reset_index()

    df=df.drop('index', axis=1)

    return df
```

```python
# data_wide2long
data_wide2long()
```

```python
# 3) Rajoutez 3 colonnes "Opérateur", "Type Couverture" et "Réseau"
def data_add3columns():
    
    df = data_wide2long()
    #df.rename(columns={7: "Variable"}, errors="raise")
    #print(df.columns)
    df["Opérateur"]=df["variable"].str.extract(r"(.*)\sCouverture")
    df["Opérateur"]=df["Opérateur"].replace({"Par au moins un opérateur":"Tout Opérateur"})
    df["Type Couverture"]=df["variable"].str.extract(r"(C.*)\s[0-9]G")
    df["Réseau"]=df["variable"].str.extract(r"([0-9]G)")
    
    return df
```

```python
# 6) Pour quels statut et opérateur a-t-on la couverture la plus faible ?
def data_get_worst_statut_operateur():
    df = data_join_geo()
    df['value']=df['value'].astype('float64')
    df=df.groupby(['Statut','Opérateur']).agg({'value':'mean'}). \
        idxmin().values[0]
    worst_statut_operateur = df
    return worst_statut_operateur
```
